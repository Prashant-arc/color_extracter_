```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
```

```python
image = matplotlib.image.imread('fruits.jpg')
```

```python
image.shape
```

```python
x = image.reshape(-1,3)
x.shape
```

```python
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(x[:,0], x[:,1], x[:,2])
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=3)
kmeans.fit(x)
```

```python
kmeans.cluster_centers_
```

```python
from PIL import Image, ImageDraw
import numpy as np

def create_color_pallete(dominant_color,pallete_size = (300,50)):
  palette = Image.new('RGB',pallete_size)
  draw =ImageDraw.Draw(palette)

  swatch_width = pallete_size[0] // len(dominant_color)

  for i, color in enumerate(dominant_color):
    draw.rectangle([i*swatch_width,0,(i+1)*swatch_width,pallete_size[1]],fill=tuple(color))

    return palette
```

```python
create_color_pallete(kmeans.cluster_centers_.astype(int))
#
```
